```python
import pandas as pd

# Creating the dataset directly inside Jupyter
data = {
    "Name": ["Ravi", "Priya", "Arjun", "Sara", "Tom"],
    "Age": [18, None, 17, 18, 19],
    "Marks": [85, 92, None, 78, None],
    "City": ["Hyderabad", None, "Bangalore", "Chennai", "Hyderabad"]
}

df = pd.DataFrame(data)

# Save CSV file
df.to_csv("students.csv", index=False)

print("students.csv file created successfully!")
df
```

```python
# Load the dataset again to show original data
df = pd.read_csv("students.csv")
print("Original Data:")
display(df)

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Marks'].fillna(df['Marks'].median(), inplace=True)
df['City'].fillna("Unknown", inplace=True)

print("After Handling Missing Values:")
display(df)
```

```python
df.drop("City", axis=1, inplace=True)

print("After Dropping City Column:")
display(df)

# Save cleaned data
df.to_csv("cleaned_students.csv", index=False)
print("✔ cleaned_students.csv saved!")
```
